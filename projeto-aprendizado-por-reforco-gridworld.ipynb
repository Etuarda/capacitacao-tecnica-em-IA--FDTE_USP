```python


# --- Instalação (Colab) ---
!pip -q install "gymnasium[toy_text]" imageio

# --- Imports ---
import numpy as np
import matplotlib.pyplot as plt
import gymnasium as gym
import imageio
from dataclasses import dataclass
from typing import Dict, Tuple, List, Optional
from IPython.display import Image, display

np.set_printoptions(precision=3, suppress=True)

# ============================================================
# 1) UTILITÁRIOS (ε-greedy, decays, plots)
# ============================================================

def rng_from_seed(seed: int) -> np.random.Generator:
    return np.random.default_rng(seed)

def epsilon_greedy(Q: np.ndarray, state: int, epsilon: float, rng: np.random.Generator) -> int:
    """ε-greedy com desempate aleatório no argmax."""
    if rng.random() < epsilon:
        return int(rng.integers(Q.shape[1]))
    row = Q[state]
    m = np.max(row)
    best = np.flatnonzero(row == m)
    return int(rng.choice(best))

def eps_constant(eps: float, ep: int, n_episodes: int) -> float:
    return float(eps)

def eps_linear_decay(eps_start: float, eps_end: float, ep: int, n_episodes: int) -> float:
    t = ep / max(1, n_episodes - 1)
    return float(eps_start + t * (eps_end - eps_start))

def moving_average(x: np.ndarray, w: int = 50) -> np.ndarray:
    if w <= 1 or len(x) < w:
        return x
    return np.convolve(x, np.ones(w)/w, mode="valid")

def plot_curves(logs: Dict[str, np.ndarray], title: str, ma_window: int = 200) -> None:
    r = logs["reward"]
    s = logs["steps"]

    plt.figure()
    plt.plot(r, alpha=0.30, label="raw")
    if len(r) >= ma_window:
        r_ma = moving_average(r, ma_window)
        plt.plot(range(ma_window-1, len(r)), r_ma, label=f"MA({ma_window})")
    plt.title(f"{title} — Recompensa total por episódio")
    plt.xlabel("Episódio")
    plt.ylabel("Recompensa")
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(s, alpha=0.30, label="raw")
    if len(s) >= ma_window:
        s_ma = moving_average(s, ma_window)
        plt.plot(range(ma_window-1, len(s)), s_ma, label=f"MA({ma_window})")
    plt.title(f"{title} — Passos por episódio")
    plt.xlabel("Episódio")
    plt.ylabel("Passos")
    plt.legend()
    plt.show()

def summarize(logs: Dict[str, np.ndarray]) -> Dict[str, float]:
    out = {
        "reward_mean": float(np.mean(logs["reward"])),
        "reward_last100_mean": float(np.mean(logs["reward"][-100:])),
        "steps_mean": float(np.mean(logs["steps"])),
        "steps_last100_mean": float(np.mean(logs["steps"][-100:])),
    }
    if "success" in logs:
        out["success_rate_last100"] = float(np.mean(logs["success"][-100:]))
    return out

# ============================================================
# 2) CONFIG DE TREINO + TREINO TABULAR GENÉRICO (Gym Discrete)
# ============================================================

@dataclass
class TrainCfg:
    n_episodes: int = 5000
    alpha: float = 0.1
    gamma: float = 0.99
    eps_mode: str = "decay"   # "const" ou "decay"
    eps_const: float = 0.10
    eps_start: float = 1.0
    eps_end: float = 0.05
    seed: int = 42

def get_eps(cfg: TrainCfg, ep: int) -> float:
    if cfg.eps_mode == "const":
        return eps_constant(cfg.eps_const, ep, cfg.n_episodes)
    if cfg.eps_mode == "decay":
        return eps_linear_decay(cfg.eps_start, cfg.eps_end, ep, cfg.n_episodes)
    raise ValueError("eps_mode deve ser 'const' ou 'decay'.")

def train_q_learning(env, cfg: TrainCfg) -> Tuple[np.ndarray, Dict[str, np.ndarray]]:
    assert isinstance(env.observation_space, gym.spaces.Discrete)
    assert isinstance(env.action_space, gym.spaces.Discrete)

    rng = rng_from_seed(cfg.seed)
    nS, nA = env.observation_space.n, env.action_space.n
    Q = np.zeros((nS, nA), dtype=np.float64)

    rewards = np.zeros(cfg.n_episodes, dtype=np.float64)
    steps = np.zeros(cfg.n_episodes, dtype=np.int32)
    success = np.zeros(cfg.n_episodes, dtype=np.float64)

    for ep in range(cfg.n_episodes):
        s, _ = env.reset(seed=cfg.seed + ep)
        done = False
        ep_r = 0.0
        ep_steps = 0
        eps = get_eps(cfg, ep)

        while not done:
            a = epsilon_greedy(Q, s, eps, rng)
            s2, r, terminated, truncated, _ = env.step(a)
            done = terminated or truncated

            # terminal: sem bootstrap
            td_target = r + (0.0 if done else cfg.gamma * np.max(Q[s2]))
            Q[s, a] += cfg.alpha * (td_target - Q[s, a])

            s = s2
            ep_r += float(r)
            ep_steps += 1

        rewards[ep] = ep_r
        steps[ep] = ep_steps
        # heurística toy_text: sucesso quando recompensa total positiva
        success[ep] = 1.0 if ep_r > 0 else 0.0

    return Q, {"reward": rewards, "steps": steps, "success": success}

def train_sarsa(env, cfg: TrainCfg) -> Tuple[np.ndarray, Dict[str, np.ndarray]]:
    assert isinstance(env.observation_space, gym.spaces.Discrete)
    assert isinstance(env.action_space, gym.spaces.Discrete)

    rng = rng_from_seed(cfg.seed)
    nS, nA = env.observation_space.n, env.action_space.n
    Q = np.zeros((nS, nA), dtype=np.float64)

    rewards = np.zeros(cfg.n_episodes, dtype=np.float64)
    steps = np.zeros(cfg.n_episodes, dtype=np.int32)
    success = np.zeros(cfg.n_episodes, dtype=np.float64)

    for ep in range(cfg.n_episodes):
        s, _ = env.reset(seed=cfg.seed + ep)
        done = False
        ep_r = 0.0
        ep_steps = 0
        eps = get_eps(cfg, ep)

        a = epsilon_greedy(Q, s, eps, rng)

        while not done:
            s2, r, terminated, truncated, _ = env.step(a)
            done = terminated or truncated

            if done:
                # terminal: alvo = r
                td_target = r
                Q[s, a] += cfg.alpha * (td_target - Q[s, a])
                ep_r += float(r)
                ep_steps += 1
                break

            a2 = epsilon_greedy(Q, s2, eps, rng)
            td_target = r + cfg.gamma * Q[s2, a2]
            Q[s, a] += cfg.alpha * (td_target - Q[s, a])

            s, a = s2, a2
            ep_r += float(r)
            ep_steps += 1

        rewards[ep] = ep_r
        steps[ep] = ep_steps
        success[ep] = 1.0 if ep_r > 0 else 0.0

    return Q, {"reward": rewards, "steps": steps, "success": success}

# ============================================================
# 3) GRIDWORLD (custom) + RENDER RGB (para GIF)
# ============================================================

class GridWorld(gym.Env):
    metadata = {"render_modes": ["rgb_array"]}

    def __init__(self, size: int = 5, render_mode: Optional[str] = "rgb_array"):
        super().__init__()
        if size < 2:
            raise ValueError("size deve ser >= 2")
        self.size = size
        self.start = (0, 0)
        self.goal = (size - 1, size - 1)
        self.state = self.start
        self.render_mode = render_mode

        self.observation_space = gym.spaces.Discrete(size * size)
        self.action_space = gym.spaces.Discrete(4)

    def _state_to_index(self, st: Tuple[int, int]) -> int:
        return st[0] * self.size + st[1]

    def reset(self, seed: Optional[int] = None, options=None):
        super().reset(seed=seed)
        self.state = self.start
        return self._state_to_index(self.state), {}

    def step(self, action: int):
        x, y = self.state
        if action == 0: x = max(x - 1, 0)
        elif action == 1: y = min(y + 1, self.size - 1)
        elif action == 2: x = min(x + 1, self.size - 1)
        elif action == 3: y = max(y - 1, 0)
        else: raise ValueError("Ação inválida")

        self.state = (x, y)
        terminated = (self.state == self.goal)
        reward = 10.0 if terminated else -1.0
        truncated = False
        return self._state_to_index(self.state), reward, terminated, truncated, {}

    def render(self):
        cell = 60
        H = self.size * cell
        W = self.size * cell
        img = np.ones((H, W, 3), dtype=np.uint8) * 255

        # grade
        for i in range(self.size + 1):
            img[i*cell:i*cell+2, :, :] = 180
            img[:, i*cell:i*cell+2, :] = 180

        # goal (verde)
        gy, gx = self.goal
        img[gy*cell:(gy+1)*cell, gx*cell:(gx+1)*cell, :] = np.array([170, 255, 170], dtype=np.uint8)

        # agente (vermelho)
        ay, ax = self.state
        img[ay*cell:(ay+1)*cell, ax*cell:(ax+1)*cell, :] = np.array([255, 170, 170], dtype=np.uint8)

        return img

# ============================================================
# 4) GIFS — EXECUÇÃO GREEDY PÓS-TREINO (inclui Taxi)
# ============================================================

def run_greedy_collect_frames(env, Q: np.ndarray, max_steps: int = 200, seed: int = 123) -> Tuple[float, int, List[np.ndarray]]:
    frames: List[np.ndarray] = []
    s, _ = env.reset(seed=seed)
    total_r = 0.0

    for t in range(max_steps):
        # render frame (se suportar)
        if hasattr(env, "render"):
            frame = env.render()
            if frame is not None:
                frames.append(frame)

        a = int(np.argmax(Q[s]))
        s, r, terminated, truncated, _ = env.step(a)
        total_r += float(r)

        if terminated or truncated:
            return total_r, (t + 1), frames

    return total_r, max_steps, frames

def save_and_show_gif(frames: List[np.ndarray], filename: str, fps: int = 3):
    if len(frames) == 0:
        raise ValueError(f"Sem frames para salvar em {filename}.")
    imageio.mimsave(filename, frames, fps=fps)
    display(Image(filename=filename))

# ============================================================
# 5) ATIVIDADE 1 — EXECUÇÃO BÁSICA (3 AMBIENTES)
# ============================================================

print("============================================================")
print("ATIVIDADE 1 — Execução Básica (Q-learning vs SARSA)")
print("============================================================")

# --- GridWorld ---
env_gw = GridWorld(size=5, render_mode="rgb_array")
cfg_gw = TrainCfg(n_episodes=900, alpha=0.5, gamma=0.9, eps_mode="decay", eps_start=1.0, eps_end=0.05, seed=123)

Q_gw_q, log_gw_q = train_q_learning(env_gw, cfg_gw)
plot_curves(log_gw_q, "GridWorld — Q-learning (ε decay)", ma_window=30)
print("Resumo GridWorld Q:", summarize(log_gw_q))

Q_gw_s, log_gw_s = train_sarsa(env_gw, cfg_gw)
plot_curves(log_gw_s, "GridWorld — SARSA (ε decay)", ma_window=30)
print("Resumo GridWorld SARSA:", summarize(log_gw_s))

# GIFs GridWorld
r, st, frames = run_greedy_collect_frames(env_gw, Q_gw_q, max_steps=60, seed=123)
print(f"GridWorld greedy Q-learning: reward={r}, steps={st}")
save_and_show_gif(frames, "gridworld_qlearning.gif", fps=3)

r, st, frames = run_greedy_collect_frames(env_gw, Q_gw_s, max_steps=60, seed=123)
print(f"GridWorld greedy SARSA: reward={r}, steps={st}")
save_and_show_gif(frames, "gridworld_sarsa.gif", fps=3)

# --- FrozenLake (slippery=True) ---
env_fl = gym.make("FrozenLake-v1", is_slippery=True, render_mode="rgb_array")
cfg_fl = TrainCfg(n_episodes=15000, alpha=0.1, gamma=0.99, eps_mode="decay", eps_start=1.0, eps_end=0.05, seed=42)

Q_fl_q, log_fl_q = train_q_learning(env_fl, cfg_fl)
plot_curves(log_fl_q, "FrozenLake slippery=True — Q-learning (ε decay)", ma_window=200)
print("Resumo FrozenLake Q:", summarize(log_fl_q))

Q_fl_s, log_fl_s = train_sarsa(env_fl, cfg_fl)
plot_curves(log_fl_s, "FrozenLake slippery=True — SARSA (ε decay)", ma_window=200)
print("Resumo FrozenLake SARSA:", summarize(log_fl_s))

# GIFs FrozenLake
r, st, frames = run_greedy_collect_frames(env_fl, Q_fl_q, max_steps=80, seed=123)
print(f"FrozenLake greedy Q-learning (slippery=True): reward={r}, steps={st}")
save_and_show_gif(frames, "frozenlake_qlearning_slippery.gif", fps=3)

r, st, frames = run_greedy_collect_frames(env_fl, Q_fl_s, max_steps=80, seed=123)
print(f"FrozenLake greedy SARSA (slippery=True): reward={r}, steps={st}")
save_and_show_gif(frames, "frozenlake_sarsa_slippery.gif", fps=3)

# --- Taxi (treino sem render para performance) ---
env_tx_train = gym.make("Taxi-v3")
cfg_tx = TrainCfg(n_episodes=30000, alpha=0.1, gamma=0.99, eps_mode="decay", eps_start=1.0, eps_end=0.05, seed=7)

Q_tx_q, log_tx_q = train_q_learning(env_tx_train, cfg_tx)
plot_curves(log_tx_q, "Taxi — Q-learning (ε decay)", ma_window=200)
print("Resumo Taxi Q:", summarize(log_tx_q))

Q_tx_s, log_tx_s = train_sarsa(env_tx_train, cfg_tx)
plot_curves(log_tx_s, "Taxi — SARSA (ε decay)", ma_window=200)
print("Resumo Taxi SARSA:", summarize(log_tx_s))

# ============================================================
# 6) ATIVIDADE 2 — EXPLORAÇÃO (ε CONSTANTE vs ε DECAY)
# ============================================================

print("============================================================")
print("ATIVIDADE 2 — Exploração (ε constante vs ε decay)")
print("============================================================")

def compare_eps(env_factory, cfg_base: TrainCfg, title: str, ma_window: int):
    # ε constante
    cfg_const = TrainCfg(**{**cfg_base.__dict__, "eps_mode": "const", "eps_const": 0.10})
    env1 = env_factory()
    Qc_q, lc_q = train_q_learning(env1, cfg_const)

    env2 = env_factory()
    Qc_s, lc_s = train_sarsa(env2, cfg_const)

    # ε decay
    cfg_decay = TrainCfg(**{**cfg_base.__dict__, "eps_mode": "decay", "eps_start": 1.0, "eps_end": 0.05})
    env3 = env_factory()
    Qd_q, ld_q = train_q_learning(env3, cfg_decay)

    env4 = env_factory()
    Qd_s, ld_s = train_sarsa(env4, cfg_decay)

    plot_curves(lc_q, f"{title} — Q-learning (ε const)", ma_window=ma_window)
    plot_curves(ld_q, f"{title} — Q-learning (ε decay)", ma_window=ma_window)
    plot_curves(lc_s, f"{title} — SARSA (ε const)", ma_window=ma_window)
    plot_curves(ld_s, f"{title} — SARSA (ε decay)", ma_window=ma_window)

    print(f"[{title}] Resumos (últimos 100)")
    print("Q const :", summarize(lc_q))
    print("Q decay :", summarize(ld_q))
    print("SARSA const:", summarize(lc_s))
    print("SARSA decay:", summarize(ld_s))

# GridWorld
compare_eps(lambda: GridWorld(size=5, render_mode=None), TrainCfg(n_episodes=900, alpha=0.5, gamma=0.9, seed=101), "GridWorld", ma_window=30)
# FrozenLake
compare_eps(lambda: gym.make("FrozenLake-v1", is_slippery=True), TrainCfg(n_episodes=15000, alpha=0.1, gamma=0.99, seed=202), "FrozenLake slippery=True", ma_window=200)
# Taxi
compare_eps(lambda: gym.make("Taxi-v3"), TrainCfg(n_episodes=30000, alpha=0.1, gamma=0.99, seed=303), "Taxi", ma_window=200)

# ============================================================
# 7) ATIVIDADE 3 — ESTOCASTICIDADE (FrozenLake True vs False)
# ============================================================

print("============================================================")
print("ATIVIDADE 3 — Estocasticidade (FrozenLake slippery True vs False)")
print("============================================================")

def train_frozenlake(slippery: bool, seed: int = 42):
    env = gym.make("FrozenLake-v1", is_slippery=slippery, render_mode=None)
    cfg = TrainCfg(n_episodes=12000, alpha=0.1, gamma=0.99, eps_mode="decay", eps_start=1.0, eps_end=0.05, seed=seed)
    Qq, lq = train_q_learning(env, cfg)
    Qs, ls = train_sarsa(env, cfg)
    plot_curves(lq, f"FrozenLake slippery={slippery} — Q-learning", ma_window=200)
    plot_curves(ls, f"FrozenLake slippery={slippery} — SARSA", ma_window=200)
    print(f"Resumo slippery={slippery} Q:", summarize(lq))
    print(f"Resumo slippery={slippery} SARSA:", summarize(ls))
    return Qq, lq, Qs, ls

Q_flT_q, log_flT_q, Q_flT_s, log_flT_s = train_frozenlake(True, seed=42)
Q_flF_q, log_flF_q, Q_flF_s, log_flF_s = train_frozenlake(False, seed=42)

# ============================================================
# 8) ATIVIDADE 4 — POLÍTICAS APRENDIDAS (GridWorld/FrozenLake/Taxi)
# ============================================================

print("============================================================")
print("ATIVIDADE 4 — Política aprendida (visual + textual)")
print("============================================================")

# GridWorld — setas
def plot_gridworld_policy(Q: np.ndarray, size: int, title: str):
    arrows = {0:"↑", 1:"→", 2:"↓", 3:"←"}
    policy = np.argmax(Q, axis=1).reshape(size, size)

    plt.figure(figsize=(5,5))
    for i in range(size):
        for j in range(size):
            if (i, j) == (size-1, size-1):
                plt.text(j, size-1-i, "G", ha="center", va="center", fontsize=16)
            else:
                plt.text(j, size-1-i, arrows[int(policy[i,j])], ha="center", va="center", fontsize=16)
    for k in range(size+1):
        plt.axhline(k-0.5, color="gray", linewidth=1)
        plt.axvline(k-0.5, color="gray", linewidth=1)
    plt.xticks([]); plt.yticks([])
    plt.title(title)
    plt.show()

plot_gridworld_policy(Q_gw_q, 5, "GridWorld — Política (Q-learning)")
plot_gridworld_policy(Q_gw_s, 5, "GridWorld — Política (SARSA)")

# FrozenLake — setas (H e G fixos)
def plot_frozenlake_policy(Q: np.ndarray, slippery: bool, title: str):
    env = gym.make("FrozenLake-v1", is_slippery=slippery)
    desc = env.unwrapped.desc
    n = int(np.sqrt(env.observation_space.n))
    arrows = {0:"←", 1:"↓", 2:"→", 3:"↑"}

    pol = np.full((n, n), "", dtype=object)
    for s in range(env.observation_space.n):
        i, j = divmod(s, n)
        cell = desc[i, j].decode("utf-8")
        if cell in ["H", "G"]:
            pol[i, j] = cell
        else:
            pol[i, j] = arrows[int(np.argmax(Q[s]))]

    plt.figure(figsize=(5,5))
    for i in range(n):
        for j in range(n):
            plt.text(j, n-1-i, pol[i,j], ha="center", va="center", fontsize=16)
    for k in range(n+1):
        plt.axhline(k-0.5, color="gray", linewidth=1)
        plt.axvline(k-0.5, color="gray", linewidth=1)
    plt.xticks([]); plt.yticks([])
    plt.title(title)
    plt.show()

plot_frozenlake_policy(Q_fl_q, True,  "FrozenLake slippery=True — Política (Q-learning)")
plot_frozenlake_policy(Q_fl_s, True,  "FrozenLake slippery=True — Política (SARSA)")
plot_frozenlake_policy(Q_flF_q, False, "FrozenLake slippery=False — Política (Q-learning)")
plot_frozenlake_policy(Q_flF_s, False, "FrozenLake slippery=False — Política (SARSA)")

# Taxi — política textual (amostra) + execução greedy
action_names = {0:"Sul", 1:"Norte", 2:"Leste", 3:"Oeste", 4:"Pickup", 5:"Dropoff"}

def print_taxi_policy_samples(Q: np.ndarray, n_states: int = 25):
    env = gym.make("Taxi-v3")
    policy = np.argmax(Q, axis=1)
    for s in range(n_states):
        x, y, p, d = env.unwrapped.decode(s)
        a = int(policy[s])
        print(f"Estado {s:3d} | Taxi=({x},{y}) | Passageiro={p} | Destino={d} | Ação={action_names[a]}")

def run_taxi_greedy(Q: np.ndarray, max_steps: int = 200, seed: int = 123) -> Tuple[float, int]:
    env = gym.make("Taxi-v3")
    s, _ = env.reset(seed=seed)
    total = 0.0
    for t in range(max_steps):
        a = int(np.argmax(Q[s]))
        s, r, terminated, truncated, _ = env.step(a)
        total += float(r)
        if terminated or truncated:
            return total, (t + 1)
    return total, max_steps

print("\nTaxi — amostras de política (Q-learning):")
print_taxi_policy_samples(Q_tx_q, n_states=25)
r, st = run_taxi_greedy(Q_tx_q, max_steps=200, seed=123)
print(f"Taxi greedy Q-learning: reward={r}, steps={st}")

print("\nTaxi — amostras de política (SARSA):")
print_taxi_policy_samples(Q_tx_s, n_states=25)
r, st = run_taxi_greedy(Q_tx_s, max_steps=200, seed=123)
print(f"Taxi greedy SARSA: reward={r}, steps={st}")

# ============================================================
# 9) GIF DO TAXI (INCLUINDO SUA VERSÃO — corrigida e integrada)
# ============================================================

print("============================================================")
print("GIFs — Taxi (Q-learning e SARSA)")
print("============================================================")

env_tx_render = gym.make("Taxi-v3", render_mode="rgb_array")

r, st, frames = run_greedy_collect_frames(env_tx_render, Q_tx_q, max_steps=200, seed=123)
print(f"Taxi greedy Q-learning (para GIF): reward={r}, steps={st} | frames={len(frames)}")
save_and_show_gif(frames, "taxi_qlearning.gif", fps=3)

r, st, frames = run_greedy_collect_frames(env_tx_render, Q_tx_s, max_steps=200, seed=123)
print(f"Taxi greedy SARSA (para GIF): reward={r}, steps={st} | frames={len(frames)}")
save_and_show_gif(frames, "taxi_sarsa.gif", fps=3)

# ============================================================


print("============================================================")
print("FIM — Arquivos gerados: gridworld_*.gif, frozenlake_*.gif, taxi_*.gif")
print("============================================================")
```


1. Introdução

Este trabalho apresenta a implementação e a análise experimental de algoritmos clássicos de Aprendizado por Reforço baseados em Diferença Temporal, com foco nos métodos Q-learning e SARSA. Ambos os algoritmos foram avaliados em três ambientes discretos amplamente utilizados para estudo de aprendizado por reforço: um GridWorld determinístico, o ambiente FrozenLake-v1 (com e sem estocasticidade) e o ambiente Taxi-v3, caracterizado por estados compostos e necessidade de planejamento sequencial.

Os objetivos do trabalho são consolidar os conceitos de aprendizado TD, comparar métodos on-policy e off-policy, analisar o impacto das estratégias de exploração, avaliar o efeito da estocasticidade do ambiente e interpretar o comportamento do agente a partir da política aprendida.

2. Metodologia
2.1 Modelagem dos Ambientes

Todos os ambientes foram tratados como Processos de Decisão de Markov, assumindo que a decisão do agente depende apenas do estado atual, da ação escolhida, da dinâmica de transição do ambiente e da recompensa recebida.

O GridWorld é um ambiente determinístico e estruturado, no qual o agente se move em um grid até alcançar um estado objetivo. O FrozenLake apresenta estados seguros, buracos e um objetivo final, podendo operar de forma determinística ou estocástica. O Taxi-v3 possui um espaço de estados maior e mais complexo, representando a posição do táxi, a localização do passageiro e o destino final.

2.2 Algoritmos Utilizados

Foram implementados dois algoritmos de aprendizado por reforço tabular baseados em Diferença Temporal:

Q-learning, um método off-policy, que aprende uma política ótima independentemente da política utilizada durante a exploração.

SARSA, um método on-policy, que aprende a política efetivamente executada pelo agente durante o treinamento.

A principal diferença entre os algoritmos está no modo como o valor do próximo estado é estimado durante a atualização da função valor-ação.

2.3 Estratégia de Exploração

Ambos os algoritmos utilizaram política de exploração do tipo ε-greedy, sendo comparadas duas estratégias:

ε constante, mantendo exploração residual durante todo o treinamento;

ε decrescente, com alto valor inicial e redução gradual ao longo dos episódios, favorecendo exploração no início e estabilização no final.

2.4 Métricas de Avaliação

Foram analisadas as seguintes métricas:

recompensa total por episódio;

número de passos por episódio;

taxa de sucesso nos últimos 100 episódios;

médias globais e médias no regime final.

As métricas do regime final foram priorizadas, pois refletem melhor o comportamento após a convergência.

3. Resultados — Atividade 1 (Execução Básica)
3.1 GridWorld

No GridWorld, ambos os algoritmos convergiram rapidamente para políticas eficientes, com altas taxas de sucesso e trajetórias curtas. O Q-learning apresentou desempenho ligeiramente superior, com menor número médio de passos e maior recompensa média no regime final. Esse comportamento é esperado em ambientes determinísticos, nos quais o método off-policy tende a aproximar-se mais rapidamente da política ótima.

A execução greedy confirmou a convergência prática, com ambos os agentes alcançando o objetivo em oito passos.

3.2 FrozenLake (com estocasticidade)

No FrozenLake estocástico, o SARSA apresentou maior taxa de sucesso final em comparação ao Q-learning. Esse resultado indica maior robustez do método on-policy em ambientes com transições probabilísticas e recompensa esparsa. O Q-learning mostrou maior variabilidade, refletindo sensibilidade à estocasticidade.

Apesar das dificuldades, ambos os algoritmos foram capazes de atingir o estado objetivo em execução greedy, evidenciando aprendizado funcional.

3.3 Taxi-v3

No ambiente Taxi, ambos os algoritmos aprenderam políticas consistentes para a sequência correta de navegação, coleta e entrega do passageiro. O SARSA apresentou ligeira vantagem em taxa de sucesso, enquanto o Q-learning mostrou desempenho competitivo em número de passos. A execução greedy resultou em trajetórias curtas e retornos positivos, confirmando aprendizado adequado da tarefa.

4. Atividade 2 — Exploração

Os experimentos mostraram que:

Em ambientes simples e determinísticos, como o GridWorld, a estratégia de exploração tem impacto reduzido no desempenho final.

Em ambientes mais complexos ou estocásticos, como FrozenLake e Taxi, a estratégia de ε decrescente resultou em maior estabilidade, melhor taxa de sucesso e menor variância no regime final.

A exploração inicial mostrou-se essencial para descoberta de trajetórias viáveis, enquanto a redução gradual do ε favoreceu a consolidação da política aprendida.

5. Atividade 3 — Estocasticidade no FrozenLake

A comparação entre o FrozenLake determinístico e estocástico evidenciou o impacto direto da estocasticidade na dificuldade do problema. Sem estocasticidade, ambos os algoritmos convergiram rapidamente, com altas taxas de sucesso e poucos passos por episódio. Com estocasticidade, o desempenho caiu significativamente, sendo o SARSA mais robusto que o Q-learning.

6. Atividade 4 — Política Aprendida

No GridWorld, a política aprendida foi visualizada por setas, indicando trajetórias diretas até o objetivo.

No FrozenLake, a política evidenciou decisões mais conservadoras, especialmente no SARSA, para evitar estados terminais negativos.

No Taxi, a política foi analisada de forma textual, mostrando ações contextuais adequadas para navegação, coleta e entrega do passageiro.

A execução greedy, demonstrada por meio de GIFs, confirmou o comportamento aprendido em todos os ambientes.

7. Discussão Teórica
7.1 Instabilidade do Q-learning em Ambientes Estocásticos

O Q-learning pode apresentar instabilidade em ambientes estocásticos devido à alta variância nas transições e recompensas. Como o método aprende off-policy e sempre considera a melhor ação estimada no próximo estado, erros de estimativa podem ser amplificados, especialmente quando há incerteza na dinâmica do ambiente.

7.2 Ambiente Mais Difícil

O ambiente mais desafiador foi o FrozenLake com estocasticidade, devido à recompensa esparsa, presença de estados terminais negativos, transições probabilísticas e alto risco de falha. O GridWorld, por ser determinístico, foi o mais simples, enquanto o Taxi apresentou complexidade estrutural compensada por recompensas densas.

7.3 Papel da Exploração

A exploração é fundamental para evitar convergência prematura em políticas subótimas. Estratégias com ε decrescente equilibram exploração inicial e estabilidade final, favorecendo a convergência, especialmente em ambientes complexos ou incertos.

8. Conclusão

Os resultados confirmam o comportamento esperado dos algoritmos de Diferença Temporal. O Q-learning mostrou-se mais eficiente em ambientes determinísticos, enquanto o SARSA apresentou maior robustez em ambientes estocásticos. A estratégia de exploração ε decrescente foi determinante para melhor desempenho final e estabilidade. O trabalho demonstra, de forma prática e teórica, os trade-offs entre métodos on-policy e off-policy no Aprendizado por Reforço clássico.
